```python
import matplotlib.pyplot as plt
import seaborn as sns

# Suponiendo que ya tienes tus datos y mostPopularComp definidos

# Crear subfiguras compartiendo el eje x
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(20, 10), sharex='col')

i = 0
for col in range(2):
    for row in range(5):
        dataAux = data[(data['Price'] < 8000000) & (data['Company Name'] == mostPopularComp[i])]
        
        # Calcular el ancho de las barras (puedes ajustar este valor seg√∫n tus preferencias)
        binwidth = 50000
        
        ax = sns.histplot(x='Price', data=dataAux, hue='Company Name', palette='bright', 
                          ax=axes[row, col], binwidth=binwidth, edgecolor='black', legend=False)
        
        # Quitar etiqueta del eje x
        axes[row, col].set_xlabel('Precio')
        axes[row, col].set_ylabel('')
        # A√±adir una leyenda personalizada
        axes[row, col].legend(title='', labels=[mostPopularComp[i]])
        
        i += 1

        

fig.suptitle('Histograma de precio de autos por cada compa√±√≠a')
plt.tight_layout()
plt.show()
```

# Preparando datos para predicci√≥n

```python
data.head()
```

```python
data_4_prediction = data[['Price', 'Model Year', 'Mileage', 'Engine Capacity']]
data_4_prediction = data_4_prediction.join(pd.get_dummies(data['Company Name'], dtype=int)) 
data_4_prediction = data_4_prediction.join(pd.get_dummies(data['Location'], dtype=int)) 
data_4_prediction = data_4_prediction.join(pd.get_dummies(data['Engine Type'], dtype=int)) 
data_4_prediction = data_4_prediction.join(pd.get_dummies(data['Color'], dtype=int)) 
data_4_prediction = data_4_prediction.join(pd.get_dummies(data['Assembly'], dtype=int, drop_first=True)) # Eliminamos 'Imported' ya que es una relaci√≥n binaria
data_4_prediction = data_4_prediction.join(pd.get_dummies(data['Body Type'], dtype=int))
data_4_prediction = data_4_prediction.join(pd.get_dummies(data['Transmission Type'], dtype=int, drop_first=True)) # Eliminamos 'Automatic' pues es una relaci√≥n binaria
data_4_prediction = data_4_prediction.join(pd.get_dummies(data['Registration Status'], dtype=int).drop(['Un-Registered'], axis=1)) # Eliminamos 'Un-Registered' pues es una relaci√≥n binaria
data_4_prediction
```

```python
data_4_prediction.describe()
```

```python
# plt.figure(figsize=(40,40))
# sns.heatmap(round(data_4_prediction.corr(),1), annot=True,  cmap='YlGnBu')
```

```python
X = data_4_prediction.drop(['Price'], axis=1)
y = data_4_prediction['Price']
```

```python
# Separar los datos en dos partes: Conjunto de entrenamiento y uno de pruebas
from sklearn.model_selection import train_test_split
X_ent, X_pru, y_ent, y_pru = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X=X_ent, y=y_ent)
```

```python
predicciones = modelo.predict(X_pru)
```

```python
comparativa = {'Predicci√≥n' : predicciones, 'Valor Real': y_pru, 'Porcentaje': (predicciones-y_pru)/y_pru}
pd.DataFrame(comparativa)
```

```python
print(modelo.score(X_ent, y_ent))
print(modelo.score(X_pru, y_pru))
```

```python
from sklearn.metrics import mean_absolute_error
preds = modelo.predict(X_pru)
mean_absolute_error(y_pru, preds)
```
